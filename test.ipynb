```python
import yfinance as yf
import pandas as pd
import os

# 🔹 Step 1: Define NIFTY 50 stocks
nifty50_stocks = list(set([
    "POWERGRID.NS", "HINDUNILVR.NS", "INFY.NS", "NESTLEIND.NS", "ITC.NS", "ASIANPAINT.NS", "CIPLA.NS",
    "SUNPHARMA.NS", "SBILIFE.NS", "HDFCBANK.NS", "JSWSTEEL.NS", "ICICIBANK.NS", "NTPC.NS", "BHARTIARTL.NS",
    "HINDALCO.NS", "AXISBANK.NS", "TATAMOTORS.NS", "HDFCLIFE.NS", "BAJAJFINSV.NS", "TECHM.NS", "BRITANNIA.NS",
    "HCLTECH.NS", "TATASTEEL.NS", "ADANIPORTS.NS", "TCS.NS", "TATACONSUM.NS", "SBIN.NS", "MARUTI.NS",
    "KOTAKBANK.NS", "GRASIM.NS", "BAJFINANCE.NS", "SHRIRAMFIN.NS", "RELIANCE.NS", "ULTRACEMCO.NS",
    "ADANIENT.NS", "M&M.NS", "APOLLOHOSP.NS", "DRREDDY.NS", "WIPRO.NS", "COALINDIA.NS", "TITAN.NS",
    "BEL.NS", "BPCL.NS", "HEROMOTOCO.NS", "LT.NS", "EICHERMOT.NS", "BAJAJ-AUTO.NS", "INDUSINDBK.NS",
    "TRENT.NS", "ONGC.NS"
]))

# 🔹 Step 2: Create a new folder for storing data
folder_name = "NIFTY50_Data"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# 🔹 Step 3: Download, clean, and save data for each stock
for ticker in nifty50_stocks:
    print(f"📥 Fetching data for {ticker}...")

    # Download stock data
    # df = yf.download(ticker, start="2020-03-12", end="2025-03-12")˚`
    df = yf.download(ticker, start="2025-01-10", end="2025-03-12", interval="1h")

    # Keep only OHLCV and reset index
    df = df[['Open', 'High', 'Low', 'Close', 'Volume']]
    df.reset_index(inplace=True)

    # Save to a temporary CSV
    temp_filename = f"temp_{ticker}.csv"
    df.to_csv(temp_filename, index=False)

    # Reload and clean data
    df = pd.read_csv(temp_filename)
    df.drop(index=0, inplace=True)  # Remove first row
    df.reset_index(inplace=True)
    df.drop(columns=["index"], axis=1, inplace=True)  # Remove extra index column

    # Save cleaned data inside the new folder
    final_filename = os.path.join(folder_name, f"{ticker.replace('.NS', '')}_NIFTY.csv")
    df.to_csv(final_filename, index=False)

    # Remove the temporary CSV
    os.remove(temp_filename)

    print(f"✅ Saved: {final_filename}")

print("🎉 All NIFTY 50 stock data processed successfully!")
```

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

# 🔹 Step 1: Load all CSV files
folder_name = "NIFTY50_Data"
all_files = [f for f in os.listdir(folder_name) if f.endswith(".csv")]

# 🔹 Step 2: Create an empty DataFrame to store closing prices
closing_prices = pd.DataFrame()

# 🔹 Step 3: Merge all closing prices
for file in all_files:
    stock_name = file.replace("_NIFTY.csv", "")  # Extract stock name from filename
    file_path = os.path.join(folder_name, file)
    
    df = pd.read_csv(file_path, usecols=["Datetime", "Close"])  # Read only Datetime and Close
    df.rename(columns={"Close": stock_name}, inplace=True)  # Rename 'Close' column to stock name

    # Convert 'Datetime' to Unix timestamp
    df["Datetime"] = pd.to_datetime(df["Datetime"])
    df["Datetime"] = df["Datetime"].astype(int) // 10**9  # Convert to Unix timestamp (seconds)

    if closing_prices.empty:
        closing_prices = df  # Initialize with first stock
    else:
        closing_prices = pd.merge(closing_prices, df, on="Datetime", how="outer")  # Merge on 'Datetime'

# 🔹 Step 4: Set 'Datetime' as index
closing_prices.set_index("Datetime", inplace=True)

# 🔹 Step 5: Compute the correlation matrix
correlation_matrix = closing_prices.corr()

# 🔹 Step 6: Extract pairs with correlation > 0.9
corr_threshold = 0.9
high_corr_pairs = (
    correlation_matrix.where((correlation_matrix > corr_threshold) & (correlation_matrix < 1.0))
    .stack()
    .reset_index()
)

high_corr_pairs.columns = ["Stock_1", "Stock_2", "Correlation"]
high_corr_pairs = high_corr_pairs.sort_values(by="Correlation", ascending=False).drop_duplicates()

# 🔹 Step 7: Save results to CSV
high_corr_pairs.to_csv("high_corr_pairs.csv", index=False)

# 🔹 Step 8: Plot heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of NIFTY 50 Closing Prices")
plt.show()

# 🔹 Step 9: Display top highly correlated pairs
print(high_corr_pairs.head(20))
```

```python
import pandas as pd
import os
from statsmodels.tsa.stattools import coint, adfuller

# 🔹 Step 1: Load high correlation pairs
high_corr_pairs = pd.read_csv("high_corr_pairs.csv")

# 🔹 Step 2: Load all closing prices
folder_name = "NIFTY50_Data"
all_files = [f for f in os.listdir(folder_name) if f.endswith(".csv")]

# Create a dictionary to store closing prices
closing_prices = {}

for file in all_files:
    stock_name = file.replace("_NIFTY.csv", "")  # Extract stock name
    file_path = os.path.join(folder_name, file)
    
    df = pd.read_csv(file_path, usecols=["Datetime", "Close"])
    df.rename(columns={"Close": stock_name}, inplace=True)
    
    # Convert to datetime only if necessary
    if not pd.api.types.is_datetime64_any_dtype(df["Datetime"]):
        df["Datetime"] = pd.to_datetime(df["Datetime"])  # Auto-detect format

    closing_prices[stock_name] = df.set_index("Datetime")  # Store in dict with 'Datetime' as index

# 🔹 Step 3: Run Cointegration & Stationarity Tests
cointegrated_pairs = []

for _, row in high_corr_pairs.iterrows():
    stock1, stock2 = row["Stock_1"], row["Stock_2"]
    
    if stock1 in closing_prices and stock2 in closing_prices:
        # Merge on Datetime
        merged_df = closing_prices[stock1].join(closing_prices[stock2], how="inner")
        
        # Drop NaN values
        merged_df.dropna(inplace=True)
        
        # Run Engle-Granger Cointegration Test
        coint_score, coint_p_value, _ = coint(merged_df[stock1], merged_df[stock2])
        
        # Run ADF Stationarity Test on Individual Stocks
        adf_stock1_p = adfuller(merged_df[stock1])[1]
        adf_stock2_p = adfuller(merged_df[stock2])[1]
        
        # Check thresholds
        if coint_p_value < 0.01:
            cointegrated_pairs.append([stock1, stock2, coint_p_value, adf_stock1_p, adf_stock2_p])

# 🔹 Step 4: Save Results
cointegrated_df = pd.DataFrame(cointegrated_pairs, 
                               columns=["Stock_1", "Stock_2", "Cointegration P-Value", "Stock1 ADF P-Value", "Stock2 ADF P-Value"])
cointegrated_df.to_csv("cointegrated_pairs.csv", index=False)

# 🔹 Step 5: Display Results
print(cointegrated_df.head(20))
```
